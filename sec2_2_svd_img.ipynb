```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./Imgs/example2.jpg')  # RGB
img = np.sum(img, axis=2) / 3
print("Shape of the image data = " + str(img.shape))  # 

plt.imshow(img)
plt.show()
```

```python
import scipy.sparse.linalg as la

def img_compress(img, k):
    u, lm, v = la.svds(img, k=k)
    img1 = u.dot(np.diag(lm)).dot(v)
    num_para = 2*k*u.shape[0]
    return img1, num_para

k=10
img1, num_para = img_compress(img, k)
print('By keeping ' + str(k) + ' singular values, data size \n is compressed from ' + str(img.size) + ' to ' + str(num_para))
print('error = ' + str(np.linalg.norm(img-img1) / np.linalg.norm(img)))
plt.imshow(img1)
plt.show()
```

```python
k=20
img1, num_para = img_compress(img, k)
print('By keeping ' + str(k) + ' singular values, data size \n is compressed from ' + str(img.size) + ' to ' + str(num_para))
print('error = ' + str(np.linalg.norm(img-img1) / np.linalg.norm(img)))
plt.imshow(img1)
plt.show()
```
